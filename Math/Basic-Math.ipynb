```python
# ========================================
# Import Libraries
# ========================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform, binom
from sympy import symbols, diff

# ========================================
# PART 1: Statistics Basics
# ========================================
print("=== PART 1: Statistics ===")
data = [2, 4, 6, 8, 10]

# Mean
mean = np.mean(data)
print("Mean:", mean)

# Median
median = np.median(data)
print("Median:", median)

# Variance
variance = np.var(data)
print("Variance:", variance)

# Standard Deviation
std_dev = np.std(data)
print("Standard Deviation:", std_dev)

# ========================================
# PART 2: Probability Distributions
# ========================================

# Normal Distribution
x = np.linspace(-3, 3, 100)
y = norm.pdf(x, 0, 1)
plt.plot(x, y)
plt.title("Normal Distribution (mean=0, std=1)")
plt.show()

# Uniform Distribution
x_u = np.linspace(0, 1, 100)
y_u = uniform.pdf(x_u, 0, 1)
plt.plot(x_u, y_u)
plt.title("Uniform Distribution")
plt.show()

# Binomial Distribution (Coin flips)
n, p = 10, 0.5
x_b = np.arange(0, n+1)
y_b = binom.pmf(x_b, n, p)
plt.bar(x_b, y_b)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.show()

# ========================================
# PART 3: Bayes Theorem
# ========================================
print("\n=== PART 3: Bayes Theorem ===")
P_D = 0.01         # Probability of disease
P_pos_D = 0.99     # Test positive if disease
P_pos_notD = 0.05  # Test positive if no disease

# Evidence
P_pos = P_pos_D*P_D + P_pos_notD*(1-P_D)

# Posterior probability
P_D_pos = (P_pos_D*P_D)/P_pos
print("Probability of disease given positive test:", P_D_pos)

# ========================================
# PART 4: Linear Algebra
# ========================================
print("\n=== PART 4: Linear Algebra ===")

# Vectors
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
dot = np.dot(v1, v2)
print("Dot product:", dot)

# Matrices & Eigenvalues
A = np.array([[2, 0], [0, 3]])
eigvals, eigvecs = np.linalg.eig(A)
print("Eigenvalues:", eigvals)
print("Eigenvectors:\n", eigvecs)

# ========================================
# PART 5: Calculus (Gradients & Derivatives)
# ========================================
print("\n=== PART 5: Derivatives & Gradients ===")
x = symbols('x')
f = x**2 + 3*x + 5
dfdx = diff(f, x)
print("Function:", f)
print("Derivative:", dfdx)

# Visualizing gradient
x_vals = np.linspace(-10, 10, 100)
y_vals = x_vals**2 + 3*x_vals + 5
grad_vals = 2*x_vals + 3  # derivative manually

plt.plot(x_vals, y_vals, label="Function f(x)")
plt.plot(x_vals, grad_vals, label="Gradient f'(x)")
plt.title("Function & Gradient")
plt.legend()
plt.show()

# ========================================
# END OF NOTEBOOK
# ========================================
```
